***Medium Loop Problems***

```python
n=int(input())
valid=0
invalid=0
amount=0
for _ in range(1,n+1):
  a=int(input())
  if a>0 and a%100==0:
    valid+=1
    amount+=a
  else:
    invalid+=1
print(valid,invalid,amount)
```

```python
n=int(input())
while n>10:
  s=0
  while n>0:
    s+=n%10
    n=n//10
  n=s
print(s)
```

```python
n,k=map(int,input().split())
v=0
tot=0
for _ in range(1,n+1):
  s=int(input())
  if s>k:
    v+=1
    tot+=s
  else:
    tot+=s
print(tot)
print(v)
```

```python
n=int(input())
low=0
normal=0
high=0
for _ in range(1,n+1):
  p=int(input())
  if p<100:
    low+=1
  elif p>=100 and p<=300:
    normal+=1
  elif p>300:
    high+=1
print("low:",low,"normal:",normal,"high:",high)
```

```python
n=int(input())
temp=n
s=0
while n>0:
  last=n%10
  s=s*10+last
  n=n//10
print(s)
if temp==s:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
n,k=map(int,input().split())
f=0
for i in range(n,k+1):
  c=0
  for j in range(1,(i//2)+1):
    if i%j==0:
      c+=1
  if c==1:
    f+=1
print(f)


```

```python
n,k=map(int,input().split())
a=0
for i in range(n,k+1):
  b=0
  for j in range(1,i):
    if i%j==0:
      b+=j

  if i==b:
    a+=1
print(a)
```

```python
n=int(input())
temp=n
b=0
while n>0:
  last=n%10
  s=1
  for i in range(1,last+1):
    s*=i
  n=n//10
  b+=s
if b==temp:
  print("strong")
else:
  print("Not strong")
```

```python
n,k=map(int,input().split())
c=0
for i in range(n,k+1):
  s=0
  current_num = i # Use a temporary variable for digit processing
  digit=len(str(current_num))
  while current_num>0:
    last=current_num%10
    s+=last
    current_num=current_num//10
  if s%digit==0:
    c+=1
print(c)
```
